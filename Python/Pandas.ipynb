```python
orders[orders["total price"].isna()]
```

```python
tot = orders["subtotal price"] - orders["discount"] + orders["Tax"]
orders["total price"] = orders["total price"].fillna(tot)
orders["total price"].isna().sum()
```

```python
sub = orders["total price"] + orders["discount"] - orders["Tax"]
orders["subtotal price"] = orders["subtotal price"].fillna(sub)
```

```python
dis = orders["subtotal price"] + orders["Tax"] - orders["total price"]
orders["discount"] = orders["discount"].fillna(dis)
```

```python
tax = orders["total price"] - orders["subtotal price"] + orders["discount"]
orders["Tax"] = orders["Tax"].fillna(tax)
```

```python
orders.loc[[342,343]]
```

```python
products["variant sku"] = products.index
```

```python
orders = orders.merge(
    products[["variant sku", "Variant Price"]],
    on="variant sku",
    how="left"
)
```

```python
orders.head()
```

```python
qty = orders["subtotal price"] // orders["Variant Price"]
orders["product qty"] = orders["product qty"].fillna(qty)
```
